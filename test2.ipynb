```python
import pandas as pd

df = pd.read_csv('dataset/businesses.csv', index_col=0, low_memory=False)
df.head(5)
```

```python
df['categories'] = df.apply(lambda x: x.categories.split(", "), axis = 1)
```

```python
results = set()
for key, value in df.iteritems():
    if key == "categories":
        for val in value:
            for v in val:
                results.add(v)
                if v == "Restaurants":
                    print(val)
```
